# NLP Tutorial - Spam Text Message Classification using NLP

## Natural Language Processing (NLP)

Natural Language Processing (NLP) is a field of Artificial Intelligence (AI) that enables computers to understand, interpret, and generate human language. It combines computational linguistics, machine learning, and deep learning techniques to process text and speech data.

### Applications of NLP
- **Machine Translation** – Google Translate, DeepL  
- **Speech Recognition** – Siri, Google Assistant, Amazon Alexa  
- **Chatbots & Virtual Assistants** – ChatGPT, IBM Watson Assistant  
- **Sentiment Analysis** – Customer feedback analysis, Social media monitoring  
- **Text Summarization** – News summarization, AI-generated abstracts  
- **Spam Detection** – Gmail spam filter, SMS spam detection  
- **Named Entity Recognition (NER)** – Identifying names, places, and organizations in text  
- **Question Answering** – AI-driven FAQ systems, Search engine snippets  
- **Text-to-Speech (TTS) & Speech-to-Text (STT)** – Google Text-to-Speech, Microsoft Azure Speech Services  
- **Optical Character Recognition (OCR)** – Google Lens, Adobe Scan for text extraction from images  
- And so much more

# **Key NLP Techniques with Examples & Advantages**

## **1. Tokenization**
**Definition:** Splitting text into smaller units, such as words or sentences.  
**Example:**  
**Input:** `"NLP is amazing!"`  
- **Word Tokenization:** `["NLP", "is", "amazing", "!"]`  
- **Sentence Tokenization:** `["NLP is amazing!"]`  

**Advantage:** Helps in breaking down large text into manageable components for further processing.

---

## **2. Stop Words Removal**
**Definition:** Removing common words that don’t add much meaning (e.g., "is," "the," "and").  
**Example:**  
**Input:** `"The cat is sitting on the mat."`  
**After Removal:** `"cat sitting mat"`  

**Advantage:** Reduces text size and processing time by eliminating unnecessary words.

---

## **3. Topic Modelling**
**Definition:** Identifying hidden topics from large text datasets.  
**Example (LDA – Latent Dirichlet Allocation):**  
A news dataset may categorize documents into topics like **Politics, Sports, Technology** based on word distribution.  

**Advantage:** Helps in organizing and analyzing large collections of text data efficiently.

---

## **4. Sentiment Analysis**
**Definition:** Detecting emotions (positive, negative, neutral) in text.  
**Example:**  
**Input:** `"I love this product! It's fantastic!"`  
**Output:** **Positive Sentiment**  

**Advantage:** Useful for analyzing customer feedback, social media monitoring, and brand reputation.

---

## **5. Named Entity Recognition (NER)**
**Definition:** Identifying names, places, dates, and organizations in text.  
**Example:**  
**Input:** `"Elon Musk is the CEO of Tesla, which is based in the USA."`  
**Output:**  
- **Person:** Elon Musk  
- **Organization:** Tesla  
- **Location:** USA  

**Advantage:** Enhances search engines, chatbots, and information retrieval by recognizing key entities.

---

## **6. Stemming & Lemmatization**
**Definition:** Converting words to their root form.  
**Example:**  
**Input:** `"running, ran, runs"`  
- **Stemming:** `"run"` (Uses heuristic rules, may be inaccurate)  
- **Lemmatization:** `"run"` (Uses a dictionary, more accurate)  

**Advantage:** Reduces vocabulary size and improves text normalization in NLP models.

---

## **7. Text Summarization**
**Definition:** Creating a short summary from a large text.  
**Example:**  
**Input:** `"Artificial Intelligence is transforming industries by automating tasks, improving efficiency, and enabling data-driven decision-making."`  
**Summarized Output:** `"AI automates tasks and improves efficiency."`  

**Advantage:** Saves time by extracting key information from long documents.

---

## **8. Keyword Extraction**
**Definition:** Extracting important words or phrases from a text.  
**Example (TF-IDF Method):**  
**Input:** `"Machine learning and deep learning are subsets of artificial intelligence."`  
**Keywords:** `["Machine learning", "deep learning", "artificial intelligence"]`  

**Advantage:** Improves search engine optimization (SEO) and information retrieval.

---

## **9. Term Frequency in NLP**
**Definition:** Measuring how often a word appears in a document (TF-IDF).  
**Example:**  
**Document:** `"Python is great for NLP. Python has many libraries for NLP."`  
- **Term Frequency of "Python"** = 2 (appears twice)  
- **Inverse Document Frequency (IDF)** lowers weight if the word appears in many documents.  

**Advantage:** Helps in ranking important words in search engines and text analysis.

---

## **10. Machine Translation**
**Definition:** Translating text from one language to another.  
**Example (Google Translate):**  
**Input:** `"Hello, how are you?"` (English)  
**Output:** `"Hola, ¿cómo estás?"` (Spanish)  

**Advantage:** Enables cross-language communication and localization of content.

---

## **11. Word Embedding**
**Definition:** Representing words as numerical vectors to capture semantic meaning.  
**Example (Word2Vec, GloVe):**  
**"King" - "Man" + "Woman" ≈ "Queen"** (Shows relationships between words in vector space)  

**Advantage:** Enhances machine learning models by understanding word relationships and context.
