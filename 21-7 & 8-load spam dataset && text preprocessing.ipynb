# 📌 How Can We Detect Spam?
### Some common spam indicators:

- Mentions of prizes: “Win”, “Claim now”, “Congratulations!”
- Urgency: “Act now!”, “Limited offer!”
- Call to action: “Click here”, “Text WIN to 12345”
- Unusual senders: Shortcodes, random email-like texts

### Whereas ham messages:

- Are personal and part of conversations.
- Do not contain suspicious links, prizes, or urgent calls to action.
